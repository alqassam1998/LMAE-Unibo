### Part 2
Once a sensor has recorded a certain number `max` of temperatures, every time it records a new temperature, it forgets the oldes temperature.

Modify the `Sensor` object as follows:

*   Create an instance method `record_temp` that takes a float value as argument and adds it to the records. If the length of the record is greater than  `max`, the oldest record is deleted
* Create an instance method  `avg_temp` that takes no arguments and returns the average temperature recorded
*   Add an object variable, shared among all the `Sensor` objects, representing `max`

Show an example of use of  `record_temp`  and  `avg_temp`

```python
class Sensor():

    def __init__(self, name, x, y, z):
      self.name = name
      self.x = x
      self.y = y
      self.z = z
      self.max = 5
      self.record = []

    def record_temp(self, data):
      if len(self.record) < self.max:
        self.record.append(data)
      else:
        self.record.append(data)
        self.record.pop(0)

    def avg_temp(self):
      return sum(self.record) / len(self.record)
```

```python
sensor_1 = Sensor('Thermocouple', 10, 10, 10)
print(sensor_1.name)
print(sensor_1.x)
print(sensor_1.y)
print(sensor_1.z)
```

```python
sensor_1.record_temp(33)
```

```python
print(sensor_1.record)
```

```python
print(sensor_1.avg_temp())
```

### Part 3
Model a collection of sensors using a dictionary.
The dictionary contain the name of the sensors as keys, and the corresponding `Sensor` object as value.



Create a function `get_averages` that given a dictionary as defined as argument returns a new dictionary with the names of the Sensors as key and their average temperature as value.
Write also che documentation comment for the function (you don't have to follow a specific format).

Create an example of dictionary and show and example of use of the `get_averages` function.

```python
sensor_dict = {
    'Ther1': Sensor('Ther1', 10, 10, 10),
    'Ther2': Sensor('Ther2', 10, 10, 10),
    'Ther3': Sensor('Ther3', 10, 10, 10),
}

sensor_dict['Ther1'].record_temp(29)
sensor_dict['Ther2'].record_temp(33)
sensor_dict['Ther3'].record_temp(22)

new_dict = {}
for key in sensor_dict:
  new_dict['Avg_' + sensor_dict[key].name] = sensor_dict[key].avg_temp()
```
