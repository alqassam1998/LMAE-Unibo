## Identities
Let's look at the difference between identities and equality

```python
l2 = [2,3]
t3 = (1, l2)
```

```python
l2 == [2,3] # l1 has the same value of this list
```

```python
l2 is [2,3] # but it is not the same object, therefore they have different identities
```

```python
l2 is t3[1] # but the second element of t2 and l1 are the same object
```

```python
[2,3] == [2,3] # here we create two objects with the same value
```

```python
[2,3] is [2,3] # but they are still two separate objects
```

```python
type([2,3]) is type([2,3])
```

## Import
Two possible ways to import the "square root" function from the math module

```python
import math # import the whole module

x = math.sqrt(9) # use them calling the name of the module first

print(x)
```

```python
from math import sqrt # import only specific parts

x = sqrt(9) # use them just through their name

print(x)
```

Be careful to not import the same module multiple times!




You can also import other objects besides functions, for example constants. And you can do multiple imports at the same time.

```python
from math import pi, e
```

Look at the Variables menu before running the next lines of code.
You can see that the constants are now in the machine as variables.

```python
print(pi) # greek pi
print(e) # Nepero's/Euler's number
# yes, I know, constants are supposed to be written in upper case :(
```
